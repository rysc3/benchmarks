## Sweeping through Ns Values
- NB = 216

````python
Ns = [30, 60, 90, 120, 150, 180, 270, 500, 560, 560.736]
speed = [1.2443e+03, 1.4197e+03, 1.4947e+03, 1.5472e+03, 1.5803e+03, 1.6054e+03, 1.6440e+03, 1.6440e+03, 1.6440e+03, 1.6440e+03]
time = [0.283, 1.75, 5.55, 12.6, 23.98, 40.75, 150, 150, 150, 150]

# Create a new figure and axis with extra space above
fig, ax1 = plt.subplots(figsize=(8, 6))

# Plot the data for GFlops
ax1.plot(Ns, speed, marker='o', linestyle='-', label='Speed (GFlops)')
ax1.set_xlabel('Ns Value (x1,000)')
ax1.set_ylabel('Speed (GFlops)', color='b')
ax1.set_title('Scaling across Ns values')

# Create a twin y-axis for time
ax2 = ax1.twinx()
ax2.plot(Ns, time, marker='s', linestyle='-', color='g', label='Time (minutes)')
ax2.set_ylabel('Time (minutes)', color='g')

# Set custom tick locations and labels for the two axes
speed_ticks = [0, 500, 1000, 1500, 2000]
time_ticks = [0, 10, 20, 30, 40, 50]
ax1.set_yticks(speed_ticks)
ax2.set_yticks(time_ticks)

# Combine the legends from both lines
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines + lines2, labels + labels2, loc='upper right')

# Show the plot
plt.show()
````
