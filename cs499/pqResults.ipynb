```python
import matplotlib.pyplot as plt
from matplotlib.ticker import FormatStrFormatter

# NB = 256
p_256 = [2, 3, 2, 4, 5, 8, 7]
q_256 = [8, 5, 7, 4, 3, 2, 2]
g_256 = [9.4977, 9.1641, 6.7772, 9.5036, 9.0142, 8.3929, 7.8243]

# NB = 432
p_432 = [2, 3, 4, 5, 7, 8]
q_432 = [8, 5, 4, 3, 2, 2]
g_432 = [9.2422, 8.7862, 9.1243, 8.6288, 7.4671, 6.8735]

# NB = 208
p_208 = [2,2,3,4,5,7,8]
q_208 = [8,7,5,4,3,2,2]
g_208 = [9.8532,8.8425,9.3747,8.5740,9.2078,8.7791,7.1468]

# NB = 352
p_352 = [2,2,3,4,5,7,8]
q_352 = [8,7,5,4,3,2,2]
g_352 = [8.7406,8.5246,8.5188,7.1518,8.4867,7.7814,8.5275]

# NB = 288
p_288 = [2,2,3,4,5,8,7,5]
q_288 = [7,8,5,4,3,2,2,3]
g_288 = [8.6668,8.8719,9.1777,9.6207,8.8477,7.5377,6.8781,8.6288]

plt.scatter(p_256, q_256, c=g_256, cmap='viridis', marker='o', label='NB = 256')
plt.scatter(p_432, q_432, c=g_432, cmap='viridis', marker='s', label='NB = 432')
plt.scatter(p_208, q_208, c=g_208, cmap='viridis', marker='d', label='NB = 208')
plt.scatter(p_352, q_352, c=g_352, cmap='viridis', marker='p', label='NB = 352')
plt.scatter(p_288, q_288, c=g_288, cmap='viridis', marker='h', label='NB = 288')

plt.xlim(1,10)
plt.ylim(1, 10)
plt.gca().xaxis.set_major_formatter(FormatStrFormatter('%.3f'))
plt.gca().yaxis.set_major_formatter(FormatStrFormatter('%.3f'))

plt.xlabel('GFlops')
plt.ylabel('Values')

plt.colorbar(label='Output (Z)')

# Add a legend to distinguish between NB values
plt.legend()

plt.show()
```
